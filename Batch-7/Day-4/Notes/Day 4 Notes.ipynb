```python
#break is used for breaking the loop 

```

```python
for i in range(1,10):
    print(i)
    if i== 5:
        break
```

```python
# Continue - is used for skipping the closed Loop 
for i in range(1,10):
    print(i)
    if i== 5:
        continue
    print("HEY I AM SAI And teaching u guys python ")
```

# File Handling ?

```python
# Create a File, Open that File and Write to that file 
file = open("letsupgrade.txt","w")
file.write("Hey Guys I Hope you will meet me on Monday")
file.close()
```

```python
# Read fromt the File we created 

file = open("letsupgrade.txt","r")
fileData = file.read()
print(fileData)
file.close()
```

```python
# Lets Append data to the File 

file = open("letsupgrade.txt","a")
file.write("\nRate today class as compared to rest out of 5")
file.close()
```

```python
# For reading and writing together we use r+
file = open("letsupgrade.txt","r+")
file.write("Hey Guys !!!")
file.close()
```
